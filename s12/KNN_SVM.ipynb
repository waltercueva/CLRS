#SVM

```python
import numpy as np
import pandas as pd
#modelo de clasificacion de SVM
from sklearn.svm import SVC
#para dividir el dataset en entrenamiento y test
from sklearn.model_selection import train_test_split
#para obtener la matriz de confusion
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.preprocessing import StandardScaler
escalado=StandardScaler()
datosXesc=escalado.fit_transform(datosX)
np.random.seed(2021)
X_e,X_p,Y_e,Y_p=train_test_split(datosXesc,datosY,test_size=0.3)
#probando con los vecinos mas cercanos
modelo=SVC(kernel="linear")
#El ajuste de modelo de entrenamiento de KNN
modelo.fit(X_e,Y_e)
y_predict=modelo.predict(X_p)
m=confusion_matrix(Y_p,y_predict)
m
```

```python
#precision para los compradores
100*m[0,0]/(m[0,0]+m[0,1])
```

```python
#precision para los no compradores
100*m[1,1]/(m[1,0]+m[1,1])
```
